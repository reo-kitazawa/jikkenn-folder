```python
#一度自分のフォルダにダウンロードしておけば2度やる必要はない
# word2vec の訓練済モデルを入手（ダウンロードに多少時間がかかります）
!wget http://www.cis.twcu.ac.jp/~asakawa/2017jpa/2017Jul_jawiki-wakati_neologd_hid200_win20_neg20_cbow.bin.gz
```

```python
#入手したデータを読み込む
#かなり時間がかかります(環境にもよりますが2分程度）
#こちらは接続のたびにやり直す必要があります
from gensim.models import KeyedVectors
w2v = KeyedVectors.load_word2vec_format('2017Jul_jawiki-wakati_neologd_hid200_win20_neg20_cbow.bin.gz',
                                                encoding='utf-8',
                                                unicode_errors='replace',
                                                binary=True)
```

```python
w2v.similarity('渋谷', '新宿')
```

```python
w = w2v.most_similar('渋谷', topn=100)

w = [str(i) for i in w]
print('/'.join(w))
```

```python
path_w = '/content/drive/MyDrive/Simpson/京王井の頭線.txt'
with open(path_w, mode='w') as f:
    f.write('\n'.join(w))
```
